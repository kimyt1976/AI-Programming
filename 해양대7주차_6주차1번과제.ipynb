```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

#산점도 점의 개수 m = 50
m = 50
x = 8 * np.random.rand(m,1) - 4
y = 0.5 * x ** 2 + 0.5 * 2 * x + 1 + np.random.randn(m, 1)

poly_features = PolynomialFeatures(degree = 2, include_bias = False)
x_poly = poly_features.fit_transform(x)
regr = LinearRegression()
regr.fit(x_poly, y)
print('다항 회귀 모델의 점수 =', regr.score(x_poly, y).round(3))
print('다항 회귀 모델의 점수 =', regr.coef_, '절편 = ', regr.intercept_)

domain = np.linspace(-4, 4, 50).reshape(-1, 1)
domain2 = poly_features.fit_transform(domain)
# 그림의 크기 (5,3)
plt.figure(figsize=(5,3))
y_predict = regr.predict(domain2)
plt.scatter(x, y)
# 다항회귀선 색깔=yellow, 굵기=3
plt.plot(domain, y_predict, color='yellow', linewidth=3)
```
